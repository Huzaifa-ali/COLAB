```python
import qrcode
```

```python
qr = qrcode.QRCode(
    version = 1,
    box_size = 10,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    mask_pattern=None,
    border = 5
)

data = 'https://www.youtube.com/'
qr.add_data(data)
qr.make(fit=True)
img = qr.make_image(fill='black',back_color='white')
img.save('text.png')
```
