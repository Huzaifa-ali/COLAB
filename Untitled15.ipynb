```python
import re
regex = "^[A-Za-z0-9]+[-._]*[A-Za-z0-9]+@[A-Za-z-]+([.]\w{2,4})?[.]\w{2,4}$"


def validating_email_address():
  email = str(input('Enter email address : '))
  if(re.match(regex,email)):
    print(f"This email adress is valid")
  else:
    print(f"This email adress is not valid")

validating_email_address()
```
